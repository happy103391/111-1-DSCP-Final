```python
# use read.csv() to make dataset useable in this notebook
air = pd.read_csv("../input/taiwans-air-quality-data-by-hours/file.csv")
```

```python
# replace missing value to 0
air_without_missing = air.fillna(0)
air_without_missing = air_without_missing.replace('ND',0)
air_without_missing = air_without_missing.replace('NaN',0)
air_without_missing = air_without_missing.replace('-',0)
air_without_missing
```

```python
# convert the data type of some columns from "object" to "float64"
air_without_missing.dtypes
air_without_missing['PM2.5'] = air_without_missing['PM2.5'].astype(float)
air_without_missing['CO'] = air_without_missing['CO'].astype(float)
air_without_missing['O3'] = air_without_missing['O3'].astype(float)
air_without_missing['O3_8hr'] = air_without_missing['O3_8hr'].astype(float)
air_without_missing['PM10'] = air_without_missing['PM10'].astype(float)
air_without_missing['WindSpeed'] = air_without_missing['WindSpeed'].astype(float)
air_without_missing['CO_8hr'] = air_without_missing['CO_8hr'].astype(float)
air_without_missing['PM10_AVG'] = air_without_missing['PM10_AVG'].astype(float)
air_without_missing['WindSpeed'] = air_without_missing['WindSpeed'].astype(float)

# check the data type of each column
air_without_missing.dtypes
```

```python
# filt tha data we want (e.g. location we want to find out)
# the user will type in the location he/she wants to 
location = input("Please type the location you want to find:")

# filt all the locations we've just type in
air_data_filter = (air_without_missing['SiteName'] == location)
air_we_want = air_without_missing.loc[air_data_filter]
air_we_want

# users can type in the type of pollutant they want to look up or analyze
type_of_pollutants = input("Please type the air pollutant you want to analyze:")
chart_df = air_we_want[["DataCreationDate", 'SiteName', type_of_pollutants]]

# convert "DataCreationDate" from string to "DateTime" to sort the data based on the time they've been created
chart_df['DataCreationDate'] = pd.to_datetime(chart_df['DataCreationDate'])
# sort the data based on the time they've been created
chart_df_sorted = chart_df.sort_values(by='DataCreationDate', ascending=True)
# set the time as the "index" of the data, so we can use the "calplot" module later
chart_df_sorted = chart_df_sorted.set_index('DataCreationDate')

# resampling the data, from 'hour' to 'day'
chart_df_sorted = chart_df_sorted.resample('D').mean()

# using "calplot" module to plot a calendar heat map 
calander_heatmap = calplot.calplot(data = chart_df_sorted[type_of_pollutants], cmap = 'GnBu', textformat = '{:.0f}', figsize = (24, 12), suptitle = type_of_pollutants +" by Month and Year")
```
